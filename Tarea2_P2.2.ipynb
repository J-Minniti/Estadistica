```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm

# Creamos una distribución Normal para comparar con la generada por el algoritmo
nd = np.linspace(norm.ppf(0.00001),
                norm.ppf(0.99999), 100)

# Creamos una muestra aleatoria con una distribución Uniforme(0,1)
nr = 30000                             # nro. de V.A. con distribución Uniforme(0,1)
y = np.random.uniform(0.0, 1.0, nr)    # y~U(0,1)


#ndtr(x) 	Gaussian cumulative distribution function
#ndtri(y) 	Inverse of ndtr vs x Returns the argument x for which the area under the Gaussian probability density function (integrated from minus infinity to x) is equal to y.
from scipy.special import ndtri

#Siguiendo el algoritmo del Problema 2.2, evaluamos la inversa de la CDF en y~U(0,1)
x = ndtri(y)


fig = plt.figure()
num_bins = 70
# Histograma de los datos
n, bins, patches = plt.hist(x, num_bins, normed=1, facecolor='blue', alpha=0.6)

plt.plot(nd, norm.pdf(nd),'r--', lw=3, alpha=0.7, label='norm pdf')
plt.xlabel('x')
plt.ylabel('Probability(x)')
plt.legend(loc='best', frameon=False)

plt.show()
```
